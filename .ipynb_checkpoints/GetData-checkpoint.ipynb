```python
market = 'BTCUSDT'
tick_interval = '1m'

for c in ['2017','2018','2019','2020'][0:1]:
    startTime = datetime.strptime('1.1.'+ c + ' 00:00:00','%d.%m.%Y %H:%M:%S').timestamp() * 1000
    stopTime = datetime.strptime('31.12.'+ c + ' 23:59:59','%d.%m.%Y %H:%M:%S').timestamp() * 1000
    time_series = []
    while True:
        print(datetime.fromtimestamp(startTime / 1e3))
#         print(startTime)
        url = 'https://api.binance.com/api/v1/klines?symbol='+market+'&interval='+tick_interval + '&startTime=' + start + '&endTime=' +stop +'&limit=480'
        data = requests.get(url).json()
        if startTime > stopTime:
            break
            time_series+=data
            
        time_series+=data
        startTime += 480* 60000
```
