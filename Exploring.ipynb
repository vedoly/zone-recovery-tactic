```python
import pandas as pd
```

```python
df = pd.read_csv('data/BNBUSDT_2018.csv')
```

```python
df = df[ ['Open time', 'Open'] ]
df = df.rename(columns={'Open time': 'Time','Open': 'Price'})
```

```python
import datetime
```

```python
df
```

```python
for i,c in enumerate(['Year','Month' , 'Day', 'Hour' ,'Minute']):
    df[c] = df['Time'].apply(lambda x: datetime.datetime.fromtimestamp(x/1000.0).timetuple()[:5][i])
df = df.drop(columns=['Time'])
df = df[ df['Year'] == 2018]
```
